Criando uma função para separar as palavras dos tokens (desconsiderando caracteres especiais)

```python
def separa_palavras(lista_tokens):
  lista_palavras=[]
  for x in lista_tokens:
    if x.isalpha() == True: #isaplha é uma função que retorna Verdadeiro se todos os caracteres da string são letras do alfabeto
      lista_palavras.append(x)
    else:
      pass
  return lista_palavras
```

Criando uma função para normalizar o texto (remover letras maiusculas) para contabilizar a palavra somente uma vez (exemplo: Temos e temos)

```python
def normalizacao(lista_palavras):
  lista_normalizada = []
  for x in lista_palavras:
    x = x.lower()
    lista_normalizada.append(x)
  return lista_normalizada
```

```python
def avaliador(lista_teste): # recebe a lista de tupulas criada pela função cria_dados_teste
  n_palavras = len(lista_teste) # quantidade de palavras 
  acertos = 0 # contador
  for correta,errada in lista_teste: # para cada palavra correta (esquerda) e palavra errada (direita)
    palavra_corrigida = corretor(errada) # usando a nossa função do corretor para corrigir as palavras
    if palavra_corrigida == correta: # se a palavra corrigida pelo nosso corretor for igual a palavra correta da base (lado esquerdo)
      acertos +=1 # soma um ao contador 
  taxa_acerto = acertos/n_palavras # taxa de acerto é a quantidade total de palavras em que o corretor retornou resultado igual a palavra correta da base divido pelo totoal de palavras existes no arquivo (o len conta a qntd de linhas)
  taxa_acerto = round(taxa_acerto*100,2)
  print(f'Do total de {n_palavras} palavras nosso corretor acertou somente {taxa_acerto}%')
```

```python
def cria_dados_teste(nome_base):
  lista_palavras_teste=[]
  f = open(nome_base,'r',) # abrindo o arquivo de texto que contem as palavras para usar no teste
  for linha in f:
    correta,errada = linha.split() # para cada linha que existe no arquivo, separar o valor da string em duas diferentes (o layout do arquivo precisa ser igual ao da aula)
    lista_palavras_teste.append((correta,errada)) # criando uma lista de tupulas com a palavra correta na esquerda e palavra errada na direita
  f.close
  return lista_palavras_teste
```

```python
def avaliador(lista_teste,vocabulario): # recebe a lista de tupulas criada pela função cria_dados_teste
  n_palavras = len(lista_teste) # quantidade de palavras 
  acertos = 0 # contador
  desconhecida = 0
  for correta,errada in lista_teste: # para cada palavra correta (esquerda) e palavra errada (direita)
    palavra_corrigida = corretor(errada) # usando a nossa função do corretor para corrigir as palavras
    if palavra_corrigida == correta: # se a palavra corrigida pelo nosso corretor for igual a palavra correta da base (lado esquerdo)
      acertos +=1 # soma um ao contador
    else:
      desconhecida += (correta not in vocabulario)
  taxa_acerto = acertos/n_palavras # taxa de acerto é a quantidade total de palavras em que o corretor retornou resultado igual a palavra correta da base divido pelo totoal de palavras existes no arquivo (o len conta a qntd de linhas)
  taxa_desconhecida = round(desconhecida*100/n_palavras, 2)
  taxa_acerto = round(taxa_acerto*100,2)
  print(f'Do total de {n_palavras} palavras nosso corretor acertou somente {taxa_acerto}%\nDa taxa de erro {taxa_desconhecida}% são por não conhecer a palavra')
```
